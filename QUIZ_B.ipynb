```python
#import the necessary modules
import pandas as pd
import numpy as np
```

```python
#import the dataset
df = pd.read_csv(r"C:\Users\ABIOLA\Downloads\energydata_complete.csv")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
#Question 13
df = df.drop(columns =['date', 'lights'])
```

```python
scaler = MinMaxScaler()
```

```python
normalised_df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
features_df = normalised_df.drop(columns = ['Appliances'])
target_df = normalised_df['Appliances']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_df, target_df, test_size = 0.3 , random_state = 42)
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
pred_values = linear_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
mae = mean_absolute_error(y_test, pred_values)
mae
round(mae, 2)
```

```python
#Question 16
r2_score = r2_score (y_test, pred_values)
round(r2_score, 2)
```

```python
#Question 15
rmse = np.sqrt(mean_squared_error(y_test, pred_values))
rmse
round(rmse, 3)
```
