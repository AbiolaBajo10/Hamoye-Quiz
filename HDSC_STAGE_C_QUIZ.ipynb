```python
#import the neccessary modules
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv(r"C:\Users\ABIOLA\Downloads\Data_for_UCI_named.csv")
```

```python
df.head()
```

```python
#encode the categorical variables
le = LabelEncoder()
df['stabf'] = le.fit_transform(df['stabf'])
df.head()
```

```python
#assign the feature and target variables to x and y respectively
x = df.drop(columns=['stab', 'stabf'], axis = 1)
y = df['stabf']
```

```python
#split the data into an 80-20 train-test split with a random state of 1
x_test, x_train, y_test, y_train = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
#create the scaler, transform the train set(x_train, y_train) and the test set(x_test)
std = StandardScaler()
x_train_std = std.fit_transform (x_train)
x_test_std  = std.transform (x_test)
```
